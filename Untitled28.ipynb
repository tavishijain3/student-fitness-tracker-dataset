## **Step 1:** Load the Dataset

```python
import pandas as pd

# Specify the file path in your Google Drive
file_path = '/content/drive/MyDrive/fitness_recommender_dataset.csv'

# Load the file into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
print(df.head())
```

## **Step 2:** Handle Missing Values

```python
# Replace NaN in 'Chronic_Condition' with 'None'
df['Chronic_Condition'] = df['Chronic_Condition'].fillna('None')

# Check for any other missing values
df.isnull().sum()
```

## **Step 3:** Standardize Text-Based Columns

```python
# Standardize categorical columns to lowercase
df['Gender'] = df['Gender'].str.lower()
df['Dietary_Preference'] = df['Dietary_Preference'].str.lower()
df['Preferred_Exercise_Type'] = df['Preferred_Exercise_Type'].str.lower()
df['Fitness_Goal'] = df['Fitness_Goal'].str.lower()
df['Activity_Level'] = df['Activity_Level'].str.lower()
df['Preferred_Workout_Time'] = df['Preferred_Workout_Time'].str.lower()
```

## **Step 4:** Handle Categorical Data

```python
from sklearn.preprocessing import LabelEncoder

# Instantiate the label encoder
label_encoder = LabelEncoder()

# Apply label encoding to categorical columns
df['Gender'] = label_encoder.fit_transform(df['Gender'])
df['Dietary_Preference'] = label_encoder.fit_transform(df['Dietary_Preference'])
df['Activity_Level'] = label_encoder.fit_transform(df['Activity_Level'])
df['Fitness_Goal'] = label_encoder.fit_transform(df['Fitness_Goal'])
df['Preferred_Workout_Time'] = label_encoder.fit_transform(df['Preferred_Workout_Time'])
df['Preferred_Exercise_Type'] = label_encoder.fit_transform(df['Preferred_Exercise_Type'])
```

## **Step 5:** Convert Numeric Columns to Correct Format

```python
# Ensure numerical columns are in the correct format
df['Age'] = df['Age'].astype(int)
df['Height_cm'] = df['Height_cm'].astype(int)
df['Weight_kg'] = df['Weight_kg'].astype(int)
df['Daily_Step_Count'] = df['Daily_Step_Count'].astype(int)
df['Resting_Heart_Rate'] = df['Resting_Heart_Rate'].astype(int)
df['Academic_Schedule'] = df['Academic_Schedule'].astype(int)
df['Weekly_Workout_Frequency'] = df['Weekly_Workout_Frequency'].astype(int)
```

## **Step 6:** Check for Duplicates

```python
# Drop duplicate rows if any
df = df.drop_duplicates()

# Verify duplicates are removed
df.duplicated().sum()
```

## **Step 7:** Feature Engineering

```python
# Calculate BMI and add it as a new column
df['BMI'] = df['Weight_kg'] / ((df['Height_cm'] / 100) ** 2)

# Display the updated dataset with BMI
df.head()
```
